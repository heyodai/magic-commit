This is a quick notebook to show the contents of the tables in the database. The tables are:

- `commits`: the commits in the repository
- `diffs`: the file changes

```python
import pandas as pd
import sqlite3
```

```python
# read in the git_diff.db
con = sqlite3.connect("data/git_diff.db")
cur = con.cursor()
```

```python
# read in the commits table
commits = pd.read_sql_query("SELECT * FROM commits", con)
commits.head()
```

```python
# read in the diff table
diff = pd.read_sql_query("SELECT * FROM diff", con)
diff.head()
```

Here's what it looks like to perform a join on the two tables.

```python
df = pd.merge(commits, diff, on='commit_hash')
df.head()
```
